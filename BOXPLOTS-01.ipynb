```python
import numpy as np
```

```python
import matplotlib.pyplot as plt
```

# Generamos un box-plot apartir de una variable 

```python
datos= [2,2,3,2,2,4,4,6,7,7,8,9,9,9,9,9,10,2,9] #ingresamos datos
```

```python
array=np.array(datos) #Creamos array
```

```python
plt.boxplot(array)#Comando para crear el box
plt.show()#Comando para mostrar el box
# Conclusion: 1/4 parte de la cantidad de indiv., esta definido entre 2 y 2,5. 
```

```python
plt.boxplot(array)#Comando para crear el box
plt.grid(True) #Creacion de grilla
plt.show()#Comando para mostrar el box
```

```python
plt.boxplot(array)#Comando para crear el box
plt.grid(True)#Creacion de grilla
plt.yticks(np.arange(2,10,0.5))#Modificacion de grilla horizontal de 2 a 10, cada 0,5.
plt.show()#Comando para mostrar el box
```

# Creamos un boxplot con valores con valores atipicos, para ver como los representa el grafico.

```python
datos1= [2,3,2,2,4,4,6,7,7,8,9,9,9,9,9,10,2,9,20,19]
```
