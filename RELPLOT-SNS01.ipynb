```python
import matplotlib.pyplot as plt
```

```python
import seaborn as sns
```

```python
import pandas as pd
```

# BASE DE DATOS DE SEABORN (DATA FRAME)

```python
sns.get_dataset_names() #Comando para ver los nombres de los dataframe.
```

```python
pinguino = sns.load_dataset('penguins') # comando para cargar el dataset, mendiante el nombre (ej:penguis)
pinguino.head(356)
```

# Transformaremos los nombres de las variables de ingles a español

```python
pinguino.columns = ["especie","isla","longitud de pico","prof de pico","long de aleta","masa","sexo"]
pinguino.head()
```

# Creamos graficas relacionales entre entre variables cuantitativas respecto a cada individuo.

```python
relacion=sns.relplot(data=pinguino,height=7,aspect=1.7)#height y aspect ESTE COMANDO AGARRA TODAS LA VARIABLES CUANTITATIVAS. Y LAS RELACIONA CON CADA INDIVIDUO EN UNA MISMA GRAFICA.
sns.set_theme(style="darkgrid",context="paper")# style y context son comandos graficos.
relacion.set(title="GRAFICA") # TITULO
relacion.set_ylabels("MASA") # NOMBRE DEL EJE X
relacion.set_xlabels("NºPINGUINOS") # NOMBRE DEL EJE Y
plt.grid(False) #retirar la grilla de la imagen
plt.show() # para ejecutar visual
```

# Creamos una grafica relacional con 2 variables designadas 

```python
sns.relplot(data=pinguino,x="longitud de pico",y="long de aleta",color="black")
plt.show()
# se puede observar una pequeña correlacion entre las variables, varianzas parecidas entre variables,y una cohovarianza positiva tendiendo a "0"
```

# Introducimos una tercera viriable (cuantitativa) a la relacion mediente un degrade de colores. 

```python
sns.relplot(data=pinguino,x="longitud de pico",y="long de aleta",hue="masa",s=34) # incorporamos tercer variable a la grafica,con el comando "hue". 
# El comando "s", es para viariar el tamaño de los puntos.
plt.show()
```

# Introduccion de una tercera variable (categorica)a la relacion mediante colores.

```python
sns.relplot(data=pinguino,x="longitud de pico",y="long de aleta",hue="especie",s=60)
plt.show()
```

# Introduccion de una cuarta variable (categorica)a la relacion anterior mediante tipo de putos.

```python
sns.relplot(data=pinguino,x="longitud de pico",y="long de aleta",hue="isla",s=60,height=7,style="sexo")
# usamos el comando style, para diferenciar el sexo
plt.grid(False)
plt.show()
```

# Introduccion de una quinta variable (cuantitativa)a la relacion anterior mediante tamaño de putos.

```python
sns.relplot(data=pinguino,x="longitud de pico",y="long de aleta",hue="isla",s=60
            ,height=6
            ,style="sexo"
            ,size="masa"
           ,sizes=(50, 200))
# usamos el comando "size" para obtener tamaño de puntos en funcion a la masa. 
plt.grid(False)
plt.show()
```

# Grafico relacional de linea

```python
# Este tipo de graficas nos permiten ver la tendecia y la fuerza
```

```python
sns.relplot(data=pinguino,x="longitud de pico",y="long de aleta",kind="line",height=7,hue="isla")
# Usamos le comando kind para pasar a grafica de linea
plt.grid(True)
plt.show()
```

```python
sns.relplot(data=pinguino,x="longitud de pico",y="long de aleta",col="isla",s=60,height=7,style="sexo",color="g")
sns.relplot(data=pinguino,x="longitud de pico",y="long de aleta",kind="line",height=7,col="isla",color="red")
plt.show()
# Unsamos el comando "col" para dezlozar puntos en funcion a una variable categorica. solo se usara para variables categoricas
```

```python
sns.relplot(data=pinguino,x="longitud de pico",y="long de aleta",kind="line",height=7,col="isla")
plt.grid(True)
plt.show()
```

```python
sns.relplot(data=pinguino,x="sexo",y="long de aleta",kind="line",height=8,hue="isla")
plt.grid(False)
plt.show()
```
